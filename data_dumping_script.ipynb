```python
import requests

# Prepping users
file = requests.get("https://raw.githubusercontent.com/alkrasnokutskaya/DE1TermProject1/refs/heads/main/user_table.csv").text.split("\r\n")[1:]
for i, row in enumerate(file[:-1]):
    row = str(i+1) + "," + row
    row = row.split(",")
    row[1] = "\"" + row[1] + "\""
    row[2] = "\"" + row[2] + "\""
    print("(", end="")
    print(*row, sep=",", end="), ")
```

```python
# Prepping friends
file = requests.get("https://raw.githubusercontent.com/alkrasnokutskaya/DE1TermProject1/refs/heads/main/friends_table.csv").text.split("\r\n")[1:]
for row in file[:-1]:
    print(f"({row}), ", end="")
```
